```python
import os
import pybel
from collections import Counter
```

```python
output_sdf = "/slgpfs/projects/irb35/agimeno/MurD/docking_MurD/structures/SagaMurD_Frag373/rDock_results_out_smiles000/results/results.sd"
```

```python
# Generate a dictionary with the best docking score for each compound
sdf_file = pybel.readfile("sdf",  output_sdf)
docking_score_dict = {}
for compound in sdf_file:
    data_dict = compound.data
    compound_title = compound.title
    docking_score = float(data_dict["SCORE.INTER"])
    if compound_title not in docking_score_dict:
        docking_score_dict[compound_title] = docking_score
    else:
        if docking_score <= docking_score_dict[compound_title]:
            docking_score_dict[compound_title] = docking_score
```

```python
file = pybel.readfile("sd", output_sdf)
aff = 1000000
affinities = {}
for molecule in file:
    if molecule.title not in affinities:
        affinities[molecule.title] = aff
    value = float(molecule.data['SCORE.INTER'])
    if value < affinities[molecule.title]:
        affinities[molecule.title] = value
```

```python
from lazyqsar.binary.classic import ClassicBinaryClassifier
```

```python
mdl = ClassicBinaryClassifier()
```

```python
import csv
with open("../data/generated/known_hits.csv", "r") as f:
    reader = csv.reader(f)
    actives = []
    for r in reader:
        actives += [r[0]]
```

```python
import random

with open("../data/docking/full_lib/smiles.smi", "r") as f:
    reader = csv.reader(f, delimiter="\t")
    inactives = []
    for r in reader:
        inactives += [r[0]]


inactives = random.sample(inactives, 10000)
```

```python
smiles = actives + inactives
y = [1]*len(actives) + [0]*len(inactives)
```

```python
mdl.fit(smiles, y)
```

```python
import joblib

joblib.dump(mdl, "../models/known_classic.joblib")
```

```python
import os
import sys
import fnmatch
import pandas as pd

# Get current dir
try: 
    ipynb_path
except NameError: 
    ipynb_path = os.getcwd()

# Join all the files named "scores" in child dirs in a single file
df_scores_all = pd.DataFrame()

for root, dirs, files in sorted(os.walk(ipynb_path, topdown=False)):
    for name in files:
        if name == 'scores':
            file_path = os.path.join(root, name)
            dir_path = os.path.dirname(file_path)
            dir_name = dir_path.split('/')[-1]
            print(dir_name)
            scores_filename = dir_path + '/' + name
            #print(scores_filename)
            # Read scores
            df_scores = pd.read_csv(scores_filename, sep='\t', names=['title', 'score'])
            #print(df_scores)
            # Append to dataframe
            df_scores_all = pd.concat([df_scores_all, df_scores])
            
# Include smiles to dataframe
# Read smiles
smiles_file = ipynb_path + '/' + 'nonisomeric_smiles' + '.smi'
df_smiles = pd.read_csv(smiles_file, sep='\t', names=['smiles', 'title'])
#print(df_smiles)

# Merge dataframes using titles
df_scores_all = pd.merge(
    df_scores_all,
    df_smiles,
    on=['title'],
    how = 'left')
#print(df_scores_all)

# Add class column for the classifier (based on docking score)
# Define docking score threshold based on the docking scores of the fragments
df_scores_fragments_file = ipynb_path + '/' + 'scores_fragments'
df_scores_fragments = pd.read_csv(df_scores_fragments_file, sep='\t', names=['title', 'score'])
#print(df_smiles)
docking_score_threshold = df_scores_fragments['score'].min()
#docking_score_threshold = df_scores_fragments['score'].mean()
#docking_score_threshold = df_scores_fragments['score'].max()
print('docking_score_threshold:', docking_score_threshold)

# Add classes to the dataframe
df_scores_all['class'] = df_scores_all.apply(lambda df_scores_all: 1 if df_scores_all['score'] <= docking_score_threshold else 0, axis = 1)

# Export dataframe as .tsv
df_scores_all_file = ipynb_path + '/' + 'docking_scores' + '.tsv'
df_scores_all.to_csv(df_scores_all_file, sep='\t', encoding='utf-8', index=False)
print(df_scores_all)
```

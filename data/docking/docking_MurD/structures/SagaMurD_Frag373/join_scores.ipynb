# 1. Join docking scores

```python
import os
import sys
import fnmatch
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Lipinski
from rdkit.Chem import Crippen
from rdkit.Chem import Descriptors
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Get current dir
try: 
    ipynb_path
except NameError: 
    ipynb_path = os.getcwd()

# Join all the files named "scores" in child dirs in a single file
df_scores_all = pd.DataFrame()

for root, dirs, files in sorted(os.walk(ipynb_path, topdown=False)):
    for name in files:
        if name == 'scores':
            file_path = os.path.join(root, name)
            dir_path = os.path.dirname(file_path)
            dir_name = dir_path.split('/')[-1]
            print(dir_name)
            scores_filename = dir_path + '/' + name
            #print(scores_filename)
            # Read scores
            df_scores = pd.read_csv(scores_filename, sep='\t', names=['title', 'score'])
            #print(df_scores)
#             # Add column with file name
#             df_scores['split'] = dir_name
            # Append to dataframe
            df_scores_all = pd.concat([df_scores_all, df_scores])
            
# Include smiles to dataframe
# Read smiles
smiles_file = ipynb_path + '/' + 'nonisomeric_smiles' + '.smi'
df_smiles = pd.read_csv(smiles_file, sep='\t', names=['smiles', 'title'])
#print(df_smiles)

# Merge dataframes using titles
df_scores_all = pd.merge(
    df_scores_all,
    df_smiles,
    on=['title'],
    how = 'left')
#print(df_scores_all)

# Define docking score threshold based on the docking scores of the fragments
df_scores_fragments_file = ipynb_path + '/' + 'scores_fragments'
df_scores_fragments = pd.read_csv(df_scores_fragments_file, sep='\t', names=['title', 'score'])
#print(df_scores_fragments)
# Read smiles for the fragments
smiles_fragments_file = ipynb_path + '/' + 'frags' + '.smi'
df_smiles_fragments = pd.read_csv(smiles_fragments_file, sep='\t', names=['smiles', 'title'])
#print(df_smiles_fragments)
# Merge dataframes using titles
df_scores_fragments = pd.merge(
    df_scores_fragments,
    df_smiles_fragments,
    on=['title'],
    how = 'left')
#print(df_scores_fragments)
smiles_fragments_list = df_scores_fragments['smiles'].tolist()
mol_fragments_list = [Chem.MolFromSmiles(smiles) for smiles in smiles_fragments_list]
# Create a list with the descriptors to calculate
descriptor_name_list = [
    'MolWt',
    #'TPSA',
    #'LabuteASA',
    #'NumHAcceptors',
    #'NumHDonors',
    #'NumHeteroatoms',
    #'NumRotatableBonds',
    #'NumAromaticRings',
    #'RingCount',
    #'MolLogP',
    #'MolMR'
]
for descriptor_name in descriptor_name_list:
    descriptor_function = getattr(Descriptors, descriptor_name) # Get function based on the name of the descriptor
    descriptor_values_list = [round(descriptor_function(mol), 2) for mol in mol_fragments_list]
    #print(descriptor_values_list)
    df_scores_fragments[descriptor_name] = descriptor_values_list
# Calculate pred LE
df_scores_fragments['score/MolWt'] = df_scores_fragments.apply(lambda row: row['score'] / row['MolWt'], axis=1)
# Export dataframe as .tsv
df_scores_fragments_file = ipynb_path + '/' + 'docking_scores_fragments' + '.tsv'
df_scores_fragments.to_csv(df_scores_fragments_file, sep='\t', encoding='utf-8', index=False)
print(df_scores_fragments)

# Calculate MW, number of non-hydrogen atoms and predicted ligand efficiency (score / # non-hydrogen atoms)
smiles_list = df_scores_all['smiles'].tolist()
mol_list = [Chem.MolFromSmiles(smiles) for smiles in smiles_list]
# Create a list with the descriptors to calculate
descriptor_name_list = [
    'MolWt',
    #'TPSA',
    #'LabuteASA',
    #'NumHAcceptors',
    #'NumHDonors',
    #'NumHeteroatoms',
    #'NumRotatableBonds',
    #'NumAromaticRings',
    #'RingCount',
    #'MolLogP',
    #'MolMR'
]
for descriptor_name in descriptor_name_list:
    descriptor_function = getattr(Descriptors, descriptor_name) # Get function based on the name of the descriptor
    descriptor_values_list = [round(descriptor_function(mol), 2) for mol in mol_list]
    #print(descriptor_values_list)
    df_scores_all[descriptor_name] = descriptor_values_list

# # Calculate QED
# #QED = [round(Chem.QED.default(mol), 2) for mol in mol_list]
# QED = []
# for mol in mol_list:
#     try:
#         QED_value = round(Chem.QED.default(mol), 2)
#     except:
#         QED_value = np.nan
#     QED.append(QED_value)
# #print(QED)
# df_scores_all['QED'] = QED

# Calculate pred LE
df_scores_all['score/MolWt'] = df_scores_all.apply(lambda row: row['score'] / row['MolWt'], axis=1)

# Add class column for the classifier (based on docking score)
threshold_property = 'score/MolWt'
#threshold_property = 'score'
#threshold = df_scores_fragments[threshold_property].min()
threshold = df_scores_fragments[threshold_property].mean()
#threshold = df_scores_fragments[threshold_property].max()
print(threshold_property + ' threshold:', threshold)
# Add classes to the dataframe
df_scores_all['class'] = df_scores_all.apply(lambda df_scores_all: 1 if df_scores_all[threshold_property] <= threshold else 0, axis = 1)

# Export dataframe as .tsv
df_scores_all_file = ipynb_path + '/' + 'docking_scores' + '.tsv'
df_scores_all.to_csv(df_scores_all_file, sep='\t', encoding='utf-8', index=False)
print(df_scores_all)

# Plot property distributions
# Create dirs
property_comparisons_dir = ipynb_path + '/' + 'property_comparisons' + '/'
# generate the folder if it doesn't exist
try:
    os.mkdir(property_comparisons_dir)
except FileExistsError:
    pass

descriptor_name_list.append('score')
descriptor_name_list.append('score/MolWt')
for descriptor_name in descriptor_name_list:
    print(descriptor_name)
    descriptor_dir = property_comparisons_dir + descriptor_name.replace("/", "-") + '/'
    # generate the folder if it doesn't exist
    try:
        os.mkdir(descriptor_dir)
    except FileExistsError:
        pass
    
    # Density plot
    try:
        library_list = ['diverse library', 'fragments']
        for lib_idx, lib_name in enumerate(library_list):
            #print(lib_name)
            # Restrict to compounds of each model
            if lib_name == 'diverse library':
                descriptor_values_list = df_scores_all[descriptor_name]
            elif lib_name == 'fragments':
                descriptor_values_list = df_scores_fragments[descriptor_name]
            label = lib_name
            if not any(np.isnan(descriptor_values_list)):
                g = sns.distplot(descriptor_values_list, hist = False, kde = True, kde_kws = {'linewidth': 3}, label=label)
        plt.legend(prop={'size': 6}, title = 'Compound library')
        plt.suptitle(descriptor_name + ' distributions')
        plt.xlabel(descriptor_name)
        plt.ylabel('Molecule density')
        plt.savefig(descriptor_dir + descriptor_name.replace("/", "-") + '_KDE.png', format="png", dpi=300)
        plt.show()
        plt.close()
    except ValueError:
        pass
```

```python
import os
import sys
import fnmatch
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Lipinski
from rdkit.Chem import Crippen
from rdkit.Chem import Descriptors
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Get current dir
try: 
    ipynb_path
except NameError: 
    ipynb_path = os.getcwd()

# Join all the files named "scores" in child dirs in a single file
df_scores_all = pd.DataFrame()

for root, dirs, files in sorted(os.walk(ipynb_path, topdown=False)):
    for name in files:
        if name == 'accessibilities.tsv':
            file_path = os.path.join(root, name)
            dir_path = os.path.dirname(file_path)
            dir_name = dir_path.split('/')[-1]
            print(dir_name)
            scores_filename = dir_path + '/' + name
            #print(scores_filename)
            # Read scores
            df_scores = pd.read_csv(scores_filename, sep='\t')
            #print(df_scores)
            # Append to dataframe
            df_scores_all = pd.concat([df_scores_all, df_scores])
            
# Include smiles to dataframe
# Read smiles
smiles_file = ipynb_path + '/' + 'nonisomeric_smiles' + '.smi'
df_smiles = pd.read_csv(smiles_file, sep='\t', names=['smiles', 'title'])
#print(df_smiles)

# Merge dataframes using titles
df_scores_all = pd.merge(
    df_scores_all,
    df_smiles,
    on=['title'],
    how = 'left')
#print(df_scores_all)

# Define docking score threshold based on the docking scores of the fragments
df_scores_fragments_file = ipynb_path + '/' + 'scores_fragments'
df_scores_fragments = pd.read_csv(df_scores_fragments_file, sep='\t', names=['title', 'score'])
#print(df_scores_fragments)
# Read smiles for the fragments
smiles_fragments_file = ipynb_path + '/' + 'frags' + '.smi'
df_smiles_fragments = pd.read_csv(smiles_fragments_file, sep='\t', names=['smiles', 'title'])
#print(df_smiles_fragments)
# Merge dataframes using titles
df_scores_fragments = pd.merge(
    df_scores_fragments,
    df_smiles_fragments,
    on=['title'],
    how = 'left')
#print(df_scores_fragments)
smiles_fragments_list = df_scores_fragments['smiles'].tolist()
mol_fragments_list = [Chem.MolFromSmiles(smiles) for smiles in smiles_fragments_list]
# Create a list with the descriptors to calculate
descriptor_name_list = [
    'MolWt',
    #'TPSA',
    #'LabuteASA',
    #'NumHAcceptors',
    #'NumHDonors',
    #'NumHeteroatoms',
    #'NumRotatableBonds',
    #'NumAromaticRings',
    #'RingCount',
    #'MolLogP',
    #'MolMR'
]
for descriptor_name in descriptor_name_list:
    descriptor_function = getattr(Descriptors, descriptor_name) # Get function based on the name of the descriptor
    descriptor_values_list = [round(descriptor_function(mol), 2) for mol in mol_fragments_list]
    #print(descriptor_values_list)
    df_scores_fragments[descriptor_name] = descriptor_values_list
# Calculate pred LE
df_scores_fragments['score/MolWt'] = df_scores_fragments.apply(lambda row: row['score'] / row['MolWt'], axis=1)
# Export dataframe as .tsv
df_scores_fragments_file = ipynb_path + '/' + 'docking_scores_fragments' + '.tsv'
df_scores_fragments.to_csv(df_scores_fragments_file, sep='\t', encoding='utf-8', index=False)
print(df_scores_fragments)

# Calculate MW, number of non-hydrogen atoms and predicted ligand efficiency (score / # non-hydrogen atoms)
print(df_scores_all)
smiles_list = df_scores_all['smiles'].tolist()
mol_list = [Chem.MolFromSmiles(smiles) for smiles in smiles_list]
# Create a list with the descriptors to calculate
descriptor_name_list = [
    'MolWt',
    #'TPSA',
    #'LabuteASA',
    #'NumHAcceptors',
    #'NumHDonors',
    #'NumHeteroatoms',
    #'NumRotatableBonds',
    #'NumAromaticRings',
    #'RingCount',
    #'MolLogP',
    #'MolMR'
]
for descriptor_name in descriptor_name_list:
    descriptor_function = getattr(Descriptors, descriptor_name) # Get function based on the name of the descriptor
    descriptor_values_list = [round(descriptor_function(mol), 2) for mol in mol_list]
    #print(descriptor_values_list)
    df_scores_all[descriptor_name] = descriptor_values_list

# # Calculate QED
# #QED = [round(Chem.QED.default(mol), 2) for mol in mol_list]
# QED = []
# for mol in mol_list:
#     try:
#         QED_value = round(Chem.QED.default(mol), 2)
#     except:
#         QED_value = np.nan
#     QED.append(QED_value)
# #print(QED)
# df_scores_all['QED'] = QED

# Calculate pred LE
df_scores_all['score/MolWt'] = df_scores_all.apply(lambda row: row['score'] / row['MolWt'], axis=1)

# Add class column for the classifier (based on docking score)
threshold_property = 'score/MolWt'
#threshold_property = 'score'
#threshold = df_scores_fragments[threshold_property].min()
threshold = df_scores_fragments[threshold_property].mean()
#threshold = df_scores_fragments[threshold_property].max()
print(threshold_property + ' threshold:', threshold)
# Add classes to the dataframe
df_scores_all['class'] = df_scores_all.apply(lambda df_scores_all: 1 if df_scores_all[threshold_property] <= threshold else 0, axis = 1)

# Export dataframe as .tsv
df_scores_all_file = ipynb_path + '/' + 'docking_scores_and_accessibilities' + '.tsv'
df_scores_all.to_csv(df_scores_all_file, sep='\t', encoding='utf-8', index=False)
print(df_scores_all)
```

```python
import os
import sys
import fnmatch
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Lipinski
from rdkit.Chem import Crippen
from rdkit.Chem import Descriptors
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Get current dir
try: 
    ipynb_path
except NameError: 
    ipynb_path = os.getcwd()

# Join all the files named "scores" in child dirs in a single file
df_scores_all = pd.DataFrame()

for root, dirs, files in sorted(os.walk(ipynb_path, topdown=False)):
    for name in files:
        if name == 'pocket_edist.tsv':
            file_path = os.path.join(root, name)
            dir_path = os.path.dirname(file_path)
            dir_name = dir_path.split('/')[-1]
            print(dir_name)
            scores_filename = dir_path + '/' + name
            #print(scores_filename)
            # Read scores
            df_scores = pd.read_csv(scores_filename, sep='\t')
            #print(df_scores)
            # Append to dataframe
            df_scores_all = pd.concat([df_scores_all, df_scores])
            
# Include smiles to dataframe
# Read smiles
smiles_file = ipynb_path + '/' + 'nonisomeric_smiles' + '.smi'
df_smiles = pd.read_csv(smiles_file, sep='\t', names=['smiles', 'title'])
#print(df_smiles)

# Merge dataframes using titles
df_scores_all = pd.merge(
    df_scores_all,
    df_smiles,
    on=['title'],
    how = 'left')
#print(df_scores_all)

# Merge with accessibility dataframe
# Read df
df_accessibilities_file = ipynb_path + '/' + 'docking_scores_and_accessibilities' + '.tsv'
df_accessibilities = pd.read_csv(df_accessibilities_file, sep='\t')
#print(df_accessibilities)
df_scores_all = pd.merge(
    df_scores_all,
    df_accessibilities,
    on=['title', 'smiles', 'score'],
    how = 'left')
#print(df_scores_all)

# Define docking score threshold based on the docking scores of the fragments
df_scores_fragments_file = ipynb_path + '/' + 'scores_fragments'
df_scores_fragments = pd.read_csv(df_scores_fragments_file, sep='\t', names=['title', 'score'])
#print(df_scores_fragments)
# Read smiles for the fragments
smiles_fragments_file = ipynb_path + '/' + 'frags' + '.smi'
df_smiles_fragments = pd.read_csv(smiles_fragments_file, sep='\t', names=['smiles', 'title'])
#print(df_smiles_fragments)
# Merge dataframes using titles
df_scores_fragments = pd.merge(
    df_scores_fragments,
    df_smiles_fragments,
    on=['title'],
    how = 'left')
#print(df_scores_fragments)
smiles_fragments_list = df_scores_fragments['smiles'].tolist()
mol_fragments_list = [Chem.MolFromSmiles(smiles) for smiles in smiles_fragments_list]
# Create a list with the descriptors to calculate
descriptor_name_list = [
    'MolWt',
    #'TPSA',
    #'LabuteASA',
    #'NumHAcceptors',
    #'NumHDonors',
    #'NumHeteroatoms',
    #'NumRotatableBonds',
    #'NumAromaticRings',
    #'RingCount',
    #'MolLogP',
    #'MolMR'
]
for descriptor_name in descriptor_name_list:
    descriptor_function = getattr(Descriptors, descriptor_name) # Get function based on the name of the descriptor
    descriptor_values_list = [round(descriptor_function(mol), 2) for mol in mol_fragments_list]
    #print(descriptor_values_list)
    df_scores_fragments[descriptor_name] = descriptor_values_list
# Calculate pred LE
df_scores_fragments['score/MolWt'] = df_scores_fragments.apply(lambda row: row['score'] / row['MolWt'], axis=1)
# Export dataframe as .tsv
df_scores_fragments_file = ipynb_path + '/' + 'docking_scores_fragments' + '.tsv'
df_scores_fragments.to_csv(df_scores_fragments_file, sep='\t', encoding='utf-8', index=False)
print(df_scores_fragments)

# Calculate MW, number of non-hydrogen atoms and predicted ligand efficiency (score / # non-hydrogen atoms)
print(df_scores_all)
smiles_list = df_scores_all['smiles'].tolist()
mol_list = [Chem.MolFromSmiles(smiles) for smiles in smiles_list]
# Create a list with the descriptors to calculate
descriptor_name_list = [
    'MolWt',
    #'TPSA',
    #'LabuteASA',
    #'NumHAcceptors',
    #'NumHDonors',
    #'NumHeteroatoms',
    #'NumRotatableBonds',
    #'NumAromaticRings',
    #'RingCount',
    #'MolLogP',
    #'MolMR'
]
for descriptor_name in descriptor_name_list:
    descriptor_function = getattr(Descriptors, descriptor_name) # Get function based on the name of the descriptor
    descriptor_values_list = [round(descriptor_function(mol), 2) for mol in mol_list]
    #print(descriptor_values_list)
    df_scores_all[descriptor_name] = descriptor_values_list

# # Calculate QED
# #QED = [round(Chem.QED.default(mol), 2) for mol in mol_list]
# QED = []
# for mol in mol_list:
#     try:
#         QED_value = round(Chem.QED.default(mol), 2)
#     except:
#         QED_value = np.nan
#     QED.append(QED_value)
# #print(QED)
# df_scores_all['QED'] = QED

# Calculate pred LE
df_scores_all['score/MolWt'] = df_scores_all.apply(lambda row: row['score'] / row['MolWt'], axis=1)

# Add class column for the classifier (based on multiple properties)
# Threshold property 1
threshold_property_1 = 'score'
print('threshold_property_1', threshold_property_1)
#print('df_scores_fragments[threshold_property_1].min()', df_scores_fragments[threshold_property_1].min())
threshold_1 = df_scores_fragments[threshold_property_1].min()
#print('df_scores_fragments[threshold_property_1].mean()', df_scores_fragments[threshold_property_1].mean())
#threshold_1 = df_scores_fragments[threshold_property_1].mean()
#print('df_scores_fragments[threshold_property_1].max()', df_scores_fragments[threshold_property_1].max())
#threshold_1 = df_scores_fragments[threshold_property_1].max()
print(threshold_property_1 + ' threshold:', threshold_1)
# Threshold property 2
threshold_property_2 = 'accessibility'
threshold_2 = 0.4
print(threshold_property_2 + ' threshold:', threshold_2)
# Threshold property 3
threshold_property_3 = 'pocket_edist'
threshold_3 = 2
print(threshold_property_3 + ' threshold:', threshold_3)
# Add classes to the dataframe
#df_scores_all['class'] = df_scores_all.apply(lambda df_scores_all: 1 if df_scores_all[threshold_property] <= threshold else 0, axis = 1)
df_scores_all['class'] = df_scores_all.apply(lambda df_scores_all: 1 if (
                                             (df_scores_all[threshold_property_1] <= threshold_1) &
                                             (df_scores_all[threshold_property_2] <= threshold_2) &
                                             (df_scores_all[threshold_property_3] <= threshold_3)
                                            ) else 0, axis = 1)

# Export dataframe as .tsv
df_scores_all_file = ipynb_path + '/' + 'docking_scores_and_accessibilities_and_pocket_edist' + '.tsv'
df_scores_all.to_csv(df_scores_all_file, sep='\t', encoding='utf-8', index=False)
print(df_scores_all)
```

```python
import os
import sys
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Read df
df_scores_all_file = ipynb_path + '/' + 'docking_scores_and_accessibilities_and_pocket_edist' + '.tsv'
df_scores_all = pd.read_csv(df_scores_all_file, sep='\t')
#print(df_scores_all)

# Plot property distributions
# Create dirs
property_comparisons_dir = ipynb_path + '/' + 'property_comparisons' + '/'
# generate the folder if it doesn't exist
try:
    os.mkdir(property_comparisons_dir)
except FileExistsError:
    pass

# Remove NaNs
#df_scores_all = df_scores_all[df_scores_all['smiles'].notna()]
#df_scores_all = df_scores_all[df_scores_all['accessibility'].notna()]
df_scores_all = df_scores_all.dropna()

descriptor_name_list.append('score')
descriptor_name_list.append('score/MolWt')
descriptor_name_list.append('accessibility')
descriptor_name_list.append('pocket_edist')
for descriptor_name in descriptor_name_list:
    print(descriptor_name)
    descriptor_dir = property_comparisons_dir + descriptor_name.replace("/", "-") + '/'
    # generate the folder if it doesn't exist
    try:
        os.mkdir(descriptor_dir)
    except FileExistsError:
        pass
    
    # Density plot
    try:
        library_list = ['diverse library', 'fragments']
        for lib_idx, lib_name in enumerate(library_list):
            #print(lib_name)
            # Restrict to compounds of each model
            if lib_name == 'diverse library':
                if descriptor_name in df_scores_all:
                    descriptor_values_list = df_scores_all[descriptor_name]
                    label = lib_name
                    if not any(np.isnan(descriptor_values_list)):
                        g = sns.distplot(descriptor_values_list, hist = False, kde = True, kde_kws = {'linewidth': 3}, label=label)
            elif lib_name == 'fragments':
                if descriptor_name in df_scores_fragments:
                    descriptor_values_list = df_scores_fragments[descriptor_name]
                    label = lib_name
                    if not any(np.isnan(descriptor_values_list)):
                        g = sns.distplot(descriptor_values_list, hist = False, kde = True, kde_kws = {'linewidth': 3}, label=label)
        plt.legend(prop={'size': 6}, title = 'Compound library')
        plt.suptitle(descriptor_name + ' distributions')
        plt.xlabel(descriptor_name)
        plt.ylabel('Molecule density')
        plt.savefig(descriptor_dir + descriptor_name.replace("/", "-") + '_KDE.png', format="png", dpi=300)
        plt.show()
        plt.close()
    except ValueError:
        pass
```
